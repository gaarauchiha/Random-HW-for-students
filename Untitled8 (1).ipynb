```python
#1.1 Spectrogram Frequency Resolution
# Part 1

import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 10000  
f1 = 500    
f2 = 520    

# Twin
Twin = 0.05  
t = np.arange(0, Twin, 1/fs)

s1 = np.cos(2 * np.pi * f1 * t)
s2 = np.cos(2 * np.pi * f2 * t)

plt.figure(figsize=(10, 5))
plt.plot(t, s1, label='s1(t) = cos(2f1t)', color='green')
plt.plot(t, s2, label='s2(t) = cos(2f2t)', color='red')
plt.title('s1(t) and s2(t)')
plt.xlabel('Time in sec')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()

```

```python
# s = s1+s2
s = s1 + s2  

plt.figure(figsize=(10, 6))
plt.plot(t, s, label='s(t) = s1(t) + s2(t)', color='green')
plt.plot(t, s1, label='s1(t)', color='blue', alpha=0.5)
plt.plot(t, s2, label='s2(t)', color='orange', alpha=0.5)
plt.title('Combined Signal s(t) = s1(t) + s2(t)')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

```python
#Part 3

f1 = 500  
f2 = 520  
fs = 10000 
duration = 0.2  
t = np.arange(0, duration, 1/fs)  

s1 = np.cos(2 * np.pi * f1 * t)
s2 = np.cos(2 * np.pi * f2 * t)
s = s1 + s2

Twin = 0.05  
visible = False

while not visible and Twin <= 0.2:
    nwin = int(Twin * fs) 
    
    frequencies, times, Sxx = spectrogram(s, fs, nperseg=nwin)
    

    plt.figure(figsize=(10, 6))
    plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='gouraud')
    plt.title(f'Spectrogram of s(t) with Twin = {Twin:.3f} s')
    plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time [s]')
    plt.colorbar(label='Intensity [dB]')
    plt.ylim(0, 1000) 
    plt.show()
    
    visible = input("Are the two spectral lines visible? (y/n): ").strip().lower() == 'y'
    
    
    if not visible:
        Twin += 0.01

C = Twin * abs(f2 - f1)

print(f"Estimated proportionality constant C: {C:.3f}")
```

```python
# Part 4
duration = 0.2  
t = np.arange(0, duration, 1/fs) 

s1 = np.cos(2 * np.pi * f1 * t)
s2 = np.cos(2 * np.pi * f2 * t)
s = s1 + s2

# FFT
n = len(s)
S = np.fft.fft(s)
f = np.fft.fftfreq(n, 1/fs)

half_n = n // 2
S_magnitude = np.abs(S[:half_n])
f_positive = f[:half_n]

plt.figure(figsize=(10, 6))
plt.plot(f_positive, S_magnitude)
plt.title('S(f)')
plt.xlabel('Freq. in Hz')
plt.ylabel('Magnitude')
plt.xlim(400, 600) 
plt.grid(True)
plt.show()
```

```python
# 1.2 Amplitude Modulation
# Part 1
import numpy as np
import matplotlib.pyplot as plt

fc = 500  
fm = 10   
fs = 5000
duration = 0.5 
t = np.arange(0, duration, 1/fs)  

v = 1 + 2 * np.cos(2 * np.pi * fm * t)
x = v * np.cos(2 * np.pi * fc * t)

plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(t, np.cos(2 * np.pi * fc * t), label='Carrier: cos(2500t)')
plt.title('Carrier')
plt.xlabel('time')
plt.ylabel('Magnitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, v, label='Envelope: 1 + 2cos(210t)', color='green')
plt.title('Initial Signal')
plt.xlabel('time')
plt.ylabel('Magnitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, x, label='AM Signal', color='red')
plt.title('AM modulated x(t)')
plt.xlabel('time')
plt.ylabel('Magnitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fc = 500  
fs = 5000  
t = np.linspace(0, 1, fs) 

v_t = 1 + 2 * np.cos(2 * np.pi * 10 * t)

x_t = v_t * np.cos(2 * np.pi * fc * t)

plt.figure(figsize=(10, 6))
plt.plot(t, v_t, label='v(t)', color='orange', linewidth=2)
plt.plot(t, x_t, label='x(t) = v(t) * cos(2fct)', color='blue', linewidth=2)
plt.title('Components of x(t)')
plt.xlabel('Time [seconds]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.xlim([0, 1])
plt.show()
```
