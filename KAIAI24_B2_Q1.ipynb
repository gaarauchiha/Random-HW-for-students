```python
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_csv('bikers_data.csv')
```

```python
data_df.head()
```

```python
data_y = data_df['Number of bikers'] # target
data_x = data_df.drop(['Number of bikers'], axis=1) # input features
```

```python
data_x.head()
```

```python
data_y
```

```python
data_x = data_df.drop(['Number of bikers', 'Date'], axis=1)
```

```python
X = data_x.values
y = data_y.values

train_size = int(0.8 * len(X))

X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Coloumn of base with 1's
X_train = np.c_[np.ones(X_train.shape[0]), X_train]
X_test = np.c_[np.ones(X_test.shape[0]), X_test]

theta = np.linalg.inv(X_train.T @ X_train) @ (X_train.T @ y_train) #Theta

# Train predictions
y_train_pred = X_train @ theta

def mean_squared_error(y, yhat):
  return 1/len(y) * np.sum(np.square(y - yhat))

train_mse = mean_squared_error(y_train, y_train_pred)
print(f'Mean Squared Error for train is = {train_mse}')

# Test Predictions
y_test_pred = X_test @ theta

test_mse = mean_squared_error(y_test, y_test_pred)
print(f'Mean Squared Error for test is = {test_mse}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data_x = data_df.drop(['Number of bikers', 'Date'], axis=1)

x_train, x_test, Y_train, Y_test = train_test_split(data_x, data_y, test_size=0.2, random_state=100)

model = LinearRegression()
model.fit(x_train, Y_train)

predictions = model.predict(x_test)
mse = mean_squared_error(Y_test, predictions)

print(f'Mean Squared Error is = {mse}')

coefficients = model.coef_
intercept = model.intercept_
print(f'Intercept: {intercept}')
print(f'Coefficients: {coefficients}')
```
