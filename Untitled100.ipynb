```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft
from scipy.signal.windows import hann


# Probelm 2

fs = 40000  # sampling Freq.
T = 1  # Period
N = int(T * fs)  # N of signal samples
t = np.arange(0, T, 1/fs) 

mu = 5000
f0 = 1000  # This value was 1000 in inital form and we changes it to match initial condition of Problem 1
delta = 0.07

xt = np.cos(2 * np.pi * (mu * (t**2) + f0 * t))  # Transmitted signal
rt = np.cos(2 * np.pi * (mu * ((t - delta)**2) + f0 * (t - delta)))  # Received signal
mt = xt * rt  # Mixed, IF, signal

window = hann(1024)  # Hanning window
nperseg = len(window)
noverlap = nperseg // 2

# STFT
f, t_stft, Zxx = stft(xt, fs, window=window, nperseg=nperseg, noverlap=noverlap)

plt.figure(figsize=(10, 6))
plt.pcolormesh(t_stft, f, np.abs(Zxx), shading='gouraud', cmap='viridis')
plt.title('Spectrogram of Transmitted Signal')
plt.ylabel('Freq in Hz')
plt.xlabel('Time in Sec')
plt.colorbar(label='Amplitude')
plt.ylim(0, fs/2)  # To make it only One-sided
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Problem 3

delta = 0.05  # Delay

rt = np.cos(2 * np.pi * (mu * ((t - delta)**2) + f0 * (t - delta)))

f, t_stft, Zxx = stft(rt, fs, window=window, nperseg=nperseg, noverlap=noverlap)

plt.figure(figsize=(10, 6))
plt.pcolormesh(t_stft, f, np.abs(Zxx), shading='gouraud', cmap='viridis')
plt.title('Spectrogram of r(t)')
plt.ylabel('Freq. in Hz')
plt.xlabel('Time in Sec')
plt.colorbar(label='Amplitude')
plt.ylim(0, fs/2) 
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Part 4

mt = xt * rt

f, t_stft, Zxx = stft(mt, fs, window=window, nperseg=nperseg, noverlap=noverlap)

plt.figure(figsize=(10, 6))
plt.pcolormesh(t_stft, f, np.abs(Zxx), shading='gouraud', cmap='viridis')
plt.title('Spectrogram of m(t)')
plt.ylabel('Freq in Hz')
plt.xlabel('Time in Sec]')
plt.colorbar(label='Amplitude')
plt.ylim(0, fs/2)  
plt.grid(True)
plt.tight_layout()
plt.show()
```

Since multiplying transmitting signal, x(t), and r(t) results in creation of higher frequencis, although for x(t) and r(t) we didnt have Aliasing since ssampling freq. was higher than 2*maximum freq of each of those signals, here for m(t) or IF signal, since the Nyquist criteria has not been met we have ALIASING.
Note that x(t)*r(t) results in frequancies twice as large as maximum frequency of each them and for IF signal m(t), 2*f_MAX > 40000 thus we DO see aliasing.
