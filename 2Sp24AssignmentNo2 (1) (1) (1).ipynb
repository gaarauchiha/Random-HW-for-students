### In this question, you are asked to implement a dictionary (called digraph) that stores a directed graph in it.

* The following is a function that checks if a directed path exist from a node A to node B in a tree structure. See if you can use this function by modifying the code (if necessary)


<div>

```python
# Your code should go here for Part(a)

def hasDirectedPath(digraph, A, B):

    if A == B:
        return True
    if A in digraph:
        for neighbor in digraph[A]:
            if hasDirectedPath(digraph, neighbor, B):
                return True
    return False

# Constructing a sample directed graph as a dictionary
digraph = {
    'A': ['B', 'C'],
    'B': ['D','E'],
    'C': ['F','G'],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

# Testing the hasDirectedPath function
node_A = 'C'
node_B = 'D'

if hasDirectedPath(digraph, node_A, node_B):
    print(f"There is a directed path from {node_A} to {node_B}.")  # Expected output: True
else:
    print(f"There is no directed path from {node_A} to {node_B}.")  # Expected output: True
    
```

```python
# We will use the digraph in part (a) to visulate and confirm the result

import networkx as nx
import matplotlib.pyplot as plt

def render_digraph(digraph):
    G = nx.DiGraph()
    for node, neighbors in digraph.items():
        for neighbor in neighbors:
            G.add_edge(node, neighbor)

    pos = nx.spring_layout(G)
    fig, ax = plt.subplots(figsize=(8, 6))
    nx.draw(
        G, pos, ax=ax, with_labels=True, node_size=1000, node_color='lightblue',
        font_size=10, font_weight='bold', arrows=True, edge_color='gray'
    )
    plt.title("Directed Graph")
    plt.show()

# Example digraph
digraph = {
    'A': ['B', 'C'],
    'B': ['D','E'],
    'C': ['F','G'],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}
# Render the example digraph
render_digraph(digraph)
```

### Task 2

In this task, you are asked to develop a Python program using your Object-oriented design in Task 1. You should define appropriate class definitions from the design in Task 1 with required attributes and methods. Your program shoud then read the student grade information from the Excel file, calculate the aggregate final score using the grading structure above, and then assign each student a letter grade using the grading ranges given in the table above. It should print the results of these calculations in the form given below (this may not be a record from the example file):

For example,
Paul Russell: score = 98, grade = A


```python
import pandas as pd

class Student:
    def __init__(self, last_name, first_name, student_id, homework_scores, quiz_scores, midterm_score, final_exam_score):
        self.last_name = last_name
        self.first_name = first_name
        self.student_id = student_id
        self.homework_scores = homework_scores
        self.quiz_scores = quiz_scores
        self.midterm_score = midterm_score
        self.final_exam_score = final_exam_score

    def calculate_final_grade(self):
        homework_weight = 0.4
        quiz_weight = 0.2
        midterm_weight = 0.2
        final_exam_weight = 0.2

        total_homework = sum(self.homework_scores) / (10 * len(self.homework_scores))
        total_quizzes = sum(self.quiz_scores) / (25 * len(self.quiz_scores))

        weighted_score = (total_homework * homework_weight +
                          total_quizzes * quiz_weight +
                          (self.midterm_score / 100) * midterm_weight +
                          (self.final_exam_score / 100) * final_exam_weight) * 100
        return weighted_score  

    def assign_letter_grade(self):
        score = self.calculate_final_grade()
        if score >= 94:
            return 'A'
        elif score >= 90:
            return 'A-'
        elif score >= 86:
            return 'B+'
        elif score >= 83:
            return 'B'
        elif score >= 80:
            return 'B-'
        elif score >= 76:
            return 'C+'
        elif score >= 73:
            return 'C'
        elif score >= 70:
            return 'C-'
        elif score >= 66:
            return 'D+'
        elif score >= 60:
            return 'D'
        else:
            return 'F'  

class Course:
    def __init__(self):
        self.students = []

    def load_students_from_csv(self, file_path):
        data = pd.read_csv(file_path)
        
        for index, row in data.iterrows():
            homework_scores = [row[f'Hw{i} (out of 10)'] for i in range(1, 5)]
            quiz_scores = [row[f'Quiz {i} (out of 25)'] for i in range(1, 5)]
            student = Student(
                last_name=row['Last Name'],
                first_name=row['First Name'],
                student_id=row['ID'],
                homework_scores=homework_scores,
                quiz_scores=quiz_scores,
                midterm_score=row['Midterm Exam (out of 100)'],
                final_exam_score=row['Final Exam (out of 100)']
            )
            self.students.append(student)

    def calculate_all_final_grades(self):
        for student in self.students:
            final_score = student.calculate_final_grade()
            letter_grade = student.assign_letter_grade()
            print(f'{student.first_name} {student.last_name}: score = {final_score:.2f}, grade = {letter_grade}')


if __name__ == "__main__":
    course = Course()
    course.load_students_from_csv('SYST003.csv')
    course.calculate_all_final_grades()
```
