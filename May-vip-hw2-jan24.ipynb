```python
import numpy as np
import matplotlib.pyplot as plt
from thinkdsp import Wave

def x(t):
    if 0 <= t < 1:
        return 0.1 * np.cos(2 * np.pi * 400 * t)
    elif 1 <= t < 2:
        return 0.1 * np.cos(2 * np.pi * 800 * t)
    elif 2 <= t < 3:
        return 0.1 * np.cos(2 * np.pi * 1600 * t)
    elif 3 <= t < 4:
        return 0.1 * np.cos(2 * np.pi * 3200 * t)
    elif 4 <= t < 5:
        return 0.1 * np.cos(2 * np.pi * 6400 * t)
    elif 5 <= t < 6:
        return 0.1 * np.cos(2 * np.pi * 12800 * t)
    else:
        return 0

def y(t):
    return 0.1 * np.cos(2 * np.pi * 1600 * t**2)

sample_rate = 11025
duration = 6

t = np.linspace(0, duration, int(sample_rate * duration))

# Sampling
x_samples = [x(ti) for ti in t]
y_samples = [y(ti) for ti in t]

# Create WaveObjects
x_wave = Wave(x_samples, framerate=sample_rate)
y_wave = Wave(y_samples, framerate=sample_rate)


#Plot
plt.subplot(2, 1, 1)
plt.plot(x_wave.ts, x_wave.ys)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('x(t)')

plt.subplot(2, 1, 2)
plt.plot(y_wave.ts, y_wave.ys)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('y(t)')


plt.tight_layout()
plt.show()

# Play
x_wave.make_audio()
y_wave.make_audio()
```
