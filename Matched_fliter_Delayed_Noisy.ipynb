```python
import numpy as np
import matplotlib.pyplot as plt

# Load Barker sequence data
barker_data = np.load('proj_part2_example_barkerdata.npz')
barker_signal = barker_data['reception'][0, :]
fs = barker_data['fs']

# Create the matched filter by time-reversing the Barker sequence
hmf = barker_signal[::-1]

# Load the noise data
noise_data = np.load('proj_part2_example_noisedata.npz')
noise_reception = noise_data['reception'][0, :]

# Load the reception data with signal
signal_reception = barker_data['reception'][0, :]

# Convolve the reception data with the matched filter
mfout_signal = np.convolve(signal_reception, hmf) / fs
mfout_noise = np.convolve(noise_reception, hmf) / fs

# Define time vectors for plotting
trecept = np.arange(0, len(signal_reception)) / fs
tmfout_signal = np.arange(0, len(mfout_signal)) / fs
tmfout_noise = np.arange(0, len(mfout_noise)) / fs

# Plot the matched filter output for the noisy reception
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(tmfout_signal, mfout_signal)
plt.title('Matched Filter Output - Signal Reception')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(tmfout_noise, mfout_noise)
plt.title('Matched Filter Output - Noise Reception')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the new datasets
dataA = np.load('proj_part2_dataA.npz')
dataB = np.load('proj_part2_dataB.npz')

fs = dataA['fs']

receptionA = dataA['reception'][0, :]
receptionB = dataB['reception'][0, :]

# Define the time vector
trecept = np.arange(0, len(receptionB)) / fs

# Define Barker 13 signal
b13 = (1 * ((trecept >= 0) & (trecept < 5))
      - 1 * ((trecept >= 5) & (trecept < 7))
      + 1 * ((trecept >= 7) & (trecept < 9))
      - 1 * ((trecept >= 9) & (trecept < 10))
      + 1 * ((trecept >= 10) & (trecept < 11))
      - 1 * ((trecept >= 11) & (trecept < 12))
      + 1 * ((trecept >= 12) & (trecept < 13)))

# Create the matched filter by reversing the Barker 13 signal
hmf = b13[::-1]

# Convolve the receptions with the matched filter
mfout_A = np.convolve(receptionA, hmf) / fs
mfout_B = np.convolve(receptionB, hmf) / fs

# Define time vectors for plotting
tmfout_A = np.arange(0, len(mfout_A)) / fs
tmfout_B = np.arange(0, len(mfout_B)) / fs

# Find the peak and arrival time for Data A
peakindex_A = np.argmax(mfout_A)
tpeak_A = tmfout_A[peakindex_A]
peak_A = np.max(mfout_A)

# Find the peak and arrival time for Data B
peakindex_B = np.argmax(mfout_B)
tpeak_B = tmfout_B[peakindex_B]
peak_B = np.max(mfout_B)

# Determine which file contains the Barker pulse
barker_file = 'Data A' if peak_A > peak_B else 'Data B'
noise_file = 'Data B' if barker_file == 'Data A' else 'Data A'
barker_time = tpeak_A if barker_file == 'Data A' else tpeak_B

# Plot the matched filter output for the new datasets
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(tmfout_A, mfout_A)
plt.title(f'Matched Filter Output - Data A (Peak at {tpeak_A:.2f}s)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(tmfout_B, mfout_B)
plt.title(f'Matched Filter Output - Data B (Peak at {tpeak_B:.2f}s)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()

print(f'Barker pulse is in: {barker_file}')
print(f'Noise is in: {noise_file}')
print(f'Time of arrival of the Barker pulse: {barker_time:.2f} seconds')
```
